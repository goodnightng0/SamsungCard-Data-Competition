```python
import pandas as pd
data=pd.read_csv("data.csv",encoding='CP949')
```

```python
data.head()
```

```python
temp=["기혼스코어","유아자녀스코어","초등학생자녀스코어","중고생자녀스코어","대학생자녀스코어","전업주부스코어"]
for i in range(len(data)):
    if i%10000==0:
        print(i)
    for j in temp:
        if data[j][i]=="high":
            data[j][i]=3
        elif data[j][i]=="mid":
            data[j][i]=2
        else:
            data[j][i]=1
```
